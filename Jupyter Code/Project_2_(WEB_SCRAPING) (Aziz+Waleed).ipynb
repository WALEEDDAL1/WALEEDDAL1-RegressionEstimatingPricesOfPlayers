```python
from bs4 import BeautifulSoup as BS
from bs4 import BeautifulSoup
import requests
import time, os
import numpy as np 
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
def get_page(num_pages):
    url = 'https://www.fifaindex.com/players/?page={}' 
    url_Player_Name=[]
    for num in num_pages :
        if num >0:
            file_url=url.format(num)
            url_Player_Name.append(file_url)
    return url_Player_Name

num_pages = list(range(41))

F_links = get_page(num_pages)
F_links
```

```python
def List_OF_Names(soup):
    links=[]
    res=[]
    link_player=[]
    mystring='https://www.fifaindex.com'
    
    for link in soup.findAll('a',{"class": "link-player"},href=True):
        
        print(link)
        links.append(link['href'])
        
        for L in links:
            
            if L not in res:
                res.append(L)
                
                
    link_player=[mystring+ s for s in res]
    return link_player
```

```python

Final_Df = pd.DataFrame(columns=['ball skills', 'physical', 'defence', 'GK', 'mental', 'shooting', 'passing', 'hight',
                                       'wight', 'age', 'value'], index= P_Names)
```

```python

L_Page = []
First_list = []
P_Names = []
columns_name = []
ball_skills = []
physical = []
defence = []
GK = []
mental = []
shooting = []
passing = []
hight = []
wight = []
value = []
age = []

driver = webdriver.Chrome(chromedriver)


for Final_Df in F_links:
    
    First_list.append(Final_Df)
    driver.get(Final_Df)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    Hold_Names = List_OF_Names(soup)
    columns_name.append(Hold_Names)
    
    for link2 in soup.findAll('a', {"class": "link-player"}):
        if link2.string != None:
            P_Names.append(link2.string)
    
    for link in Hold_Names:
        r = requests.get(link)
        html_content = r.text
        soup2 = BS(html_content, 'lxml')
        
        for div in soup2.find_all('div', attrs={'class': "pl-3"}):
            
            
            B_Control = float(soup2.find(text='Ball Control ').findNext().text)
            Dribbling = float(soup2.find(text='Dribbling ').findNext().text)
            B_Skills = np.mean([B_Control, Dribbling])
            
            
            Acc = float(soup2.find(text='Acceleration ').findNext().text)
            Stam = float(soup2.find(text='Stamina ').findNext().text)
            bal = float(soup2.find(text='Strength ').findNext().text)
            SS = float(soup2.find(text='Balance ').findNext().text)
            Ag = float(soup2.find(text='Sprint Speed ').findNext().text)
            St = float(soup2.find(text='Agility ').findNext().text)
            Jum = float(soup2.find(text='Jumping ').findNext().text)
            B_Physical = np.mean([Acc, Stam, bal, SS, Ag, St, Jum])
            
            
            Marking = float(soup2.find(text='Marking ').findNext().text)
            Stand_Tackle = float(soup2.find(
            text='Stand Tackle ').findNext().text)
            B_Defence = np.mean([Marking, Stand_Tackle])
            
            
            Positioning = float(soup2.find(
            text='GK Positioning ').findNext().text)
            Diving = float(soup2.find(text='GK Diving ').findNext().text)
            Handling = float(soup2.find(text='GK Handling ').findNext().text)
            Kicking = float(soup2.find(text='GK Kicking ').findNext().text)
            Reflexes = float(soup2.find(text='GK Reflexes ').findNext().text)
            B_Goalkeeper = np.mean(
            [Positioning, Diving, Handling, Kicking, Reflexes])
            
            
            
            B_Aggression = float(soup2.find(
            text='Aggression ').findNext().text)
            Reactions = float(soup2.find(text='Reactions ').findNext().text)
            Att_po = float(soup2.find(text='Att. Position ').findNext().text)
            Interceptions = float(soup2.find(
            text='Interceptions ').findNext().text)
            Vision = float(soup2.find(text='Vision ').findNext().text)
            Composure = float(soup2.find(text='Composure ').findNext().text)
            B_Mental = np.mean(
            [B_Aggression, Reactions, Att_po, Interceptions, Vision, Composure])
            
            
            
            Heading = float(soup2.find(text='Heading ').findNext().text)
            power = float(soup2.find(text='Shot Power ').findNext().text)
            Finishing = float(soup2.find(text='Finishing ').findNext().text)
            long = float(soup2.find(text='Long Shots ').findNext().text)
            Curve = float(soup2.find(text='Curve ').findNext().text)
            FK = float(soup2.find(text='FK Acc. ').findNext().text)
            Penalties = float(soup2.find(text='Penalties ').findNext().text)
            Volleys = float(soup2.find(text='Volleys ').findNext().text)
            B_Shooting = np.mean(
            [Heading, power, Finishing, long, Curve, FK, Penalties, Volleys])
            
            
            
            Crossing = float(soup2.find(text='Crossing ').findNext().text)
            Short_p = float(soup2.find(text='Short Pass ').findNext().text)
            Long_p = float(soup2.find(text='Long Pass ').findNext().text)
            B_Passing = np.mean([Crossing, Short_p, Long_p])
            
            
            
            Height = soup2.find_all(
            'span', class_="data-units data-units-metric")[0].text.split()[0]
            Weight = soup2.find_all(
            'span', class_="data-units data-units-metric")[1].text.split()[0]
            if(soup2.find_all('p', class_='data-currency data-currency-euro')!=[]):
                Market_Value = ''.join([i for i in soup2.find_all('p', class_='data-currency data-currency-euro')
                                [0].findNext().text if i.isdigit()])
            else:
                Market_Value=None
            age2 = soup2.find(text='Age ').findNext().text

            ball_skills.append(B_Skills)
            physical.append(B_Physical)
            defence.append(B_Defence)
            GK.append(B_Goalkeeper)
            mental.append(B_Mental)
            shooting.append(B_Shooting)
            passing.append(B_Passing)
            wight.append(Weight)
            hight.append(Height)
            value.append(Market_Value)
            age.append(age2)
            
            
            

```

```python

Final_Df = pd.DataFrame(columns=['ball skills', 'physical', 'defence', 'GK', 'mental', 'shooting', 'passing', 'hight',
                                       'wight', 'age', 'value'], index= P_Names)
```

```python

Final_Df['ball skills'] = ball_skills
Final_Df['physical'] = physical
Final_Df['defence'] = defence
Final_Df['GK'] = GK
Final_Df['mental'] = mental
Final_Df['shooting'] = shooting
Final_Df['passing'] = passing
Final_Df['hight'] = hight
Final_Df['wight'] = wight
Final_Df['age'] = age
Final_Df['value'] = value
```
